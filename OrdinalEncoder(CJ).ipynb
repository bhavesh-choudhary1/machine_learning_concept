```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("customer.csv")
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
oe.fit(df[['gender']])
```

```python
oe.transform(df[['gender']])
```

```python
np.ndim(oe.transform(df[['gender']]))
```

```python
df['gender'].unique()
```

```python
df['gender'] = oe.transform(df[['gender']])
```
