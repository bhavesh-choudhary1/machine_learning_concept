```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passenger travelling with ')
```

```python
# extract numerical part
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# extract categorical part
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['cabin_numerical'] = df['Cabin'].str.extract('(\d+)') # captures numerical part
df['Cabin_categorical'] = df['Cabin'].str[0] # captures the first letter
```

```python
df.head()
```

```python
df['Cabin_categorical'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of the ticket number
df['ticket_num'] = df['Ticket'].apply(lambda s:s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

# extract the first part of the ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,df['ticket_cat'])
```
