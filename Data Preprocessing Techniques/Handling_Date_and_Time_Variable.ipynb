##Working with Dates

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1.extract year

```python
date['date_year'] = date['date'].dt.year
```

```python
date.head()
```

2.extract month

```python
date['date_month'] = date['date'].dt.month
```

```python
date.head()
```

3.Extract Month name

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

4.Extract Days

```python
date['date_day'] = date['date'].dt.day
```

```python
date.head()
```

5.Extract Day of the week

```python
date['date_dow'] = date['date'].dt.dayofweek
```

```python
date.head()
```

6.Day of week - name

```python
date['date_dow_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

7. is Weekend ?

```python
date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.head()
```

8. Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week
```

```python
date.head()
```

9. Extract Quater

```python
date['quater'] = date['date'].dt.quarter
```

```python
date.head()
```

10. Extract Semester

```python
date['semester'] = np.where(date['quater'].isin([1,2]),1,2)
```

```python
date.head()
```

11. Extract time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# month passed

np.round((today - date['date']) / np.timedelta64(1, 'M'),0)
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second
```

Extract Time Part

```python
time['time'] = time['date'].dt.time

time.head()
```

Time Difference

```python
today - time['date']
```

```python
# in second

(today - time['date'])/np.timedelta64(1,'s')
```

```python
# in minute

(today - time['date'])/np.timedelta64(1,'m')
```

```python
# in hour

(today - time['date'])/np.timedelta64(1,'h')
```
