```python
# 1.for locally read the csv file
import pandas as pd
```

```python
df = pd.read_csv("/content/sat_gpa.csv")
df
```

```python
#different parameter in read csv file
# 1.Seperate Parameter
df = pd.read_csv("/content/spam.csv",sep='\t',encoding='latin-1')
df
```

```python
#2. Index_col parameter
df1=pd.read_csv('/content/Placement.csv',index_col='Student_ID')
df1
```

```python
#3. Header Parameter
df1 = pd.read_csv("/content/Placement.csv",header=1)
df1
```

```python
#4. usecols parameter
df2=pd.read_csv('/content/Placement.csv',usecols=['IQ','CGPA'])
df2
```

```python
#5. Squeeze parameter
#df3=pd.read_csv('/content/Placement.csv',usecols=['IQ'],squeeze=True)
#df3
#the squeeze parameter is used in the older version of the pandas
```

```python
#5. Achieve the effect of the 'squeeze' parameter
df3 = pd.read_csv('/content/Placement.csv', usecols=['IQ'])['IQ']
df3
```

```python
#6. skiprows parameter
df4=pd.read_csv('/content/Placement.csv',skiprows=[1,2,3,4])
df4
```

```python
#7. nrows parameter
df5=pd.read_csv('/content/Placement.csv',nrows=50)
df5
```

```python
#8. Encoding Parameter
df6=pd.read_csv('/content/spam.csv',encoding='latin-1')
df6
```

```python
#9. Skip Bad Lines
df7=pd.read_csv('/content/BX-Books.csv',encoding='latin-1')
df7
```

```python
#9. Skip Bad Lines
df7=pd.read_csv('/content/BX-Books.csv',encoding='latin-1', on_bad_lines='skip') # Use on_bad_lines='skip' to skip bad lines
df7
```
