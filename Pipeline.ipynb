# Titanic Dataset without usiing pipeline

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
x_train.sample()
```

```python
y_train.head()
```

```python
df.isnull().sum()
```

```python
# applying imputation

si_age =SimpleImputer()
si_embarked = SimpleImputer(strategy='most_frequent')

x_train_age = si_age.fit_transform(x_train[['Age']])
x_train_embarked = si_embarked.fit_transform(x_train[['Embarked']])

x_test_age = si_age.transform(x_test[['Age']])
x_test_embarked = si_embarked.transform(x_test[['Embarked']])
```

```python
x_train_embarked
```

```python
# onehotencoding on Sex Embarked

ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_embarked =OneHotEncoder(sparse_output=False,handle_unknown='ignore')

x_train_sex = ohe_sex.fit_transform(x_train[['Sex']])
x_train_embarked = ohe_embarked.fit_transform(x_train[['Embarked']])

x_test_sex = ohe_sex.transform(x_test[['Sex']])
x_test_embarked = ohe_embarked.transform(x_test[['Embarked']])
```

```python
x_test_embarked
```

```python
x_train_remaining = x_train.drop(columns=['Sex','Age','Embarked'])
```

```python
x_test_remaining = x_test.drop(columns=['Sex','Age','Embarked'])
```

```python
x_train_transform = np.concatenate((x_train_remaining,x_train_age,x_train_sex,x_train_embarked),axis=1)
```

```python
x_test_transform = np.concatenate((x_test_remaining,x_test_age,x_test_sex,x_test_embarked),axis=1)
```

```python
x_train_transform.shape
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train_transform,y_train)
```

```python
y_pred = clf.predict(x_test_transform)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

# Titanic_dataset-using-pipeline

missing values --> ohe --> scaling --> feature selection --> model train using decision tree

```python
# imputation transformer

trf1 = ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
],remainder='passthrough')
```

```python
# onehotencoding

trf2 = ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore',dtype=int),[1,6])
])
```

```python
#scaling

trf3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
#feature Selection

trf4 = SelectKBest(score_func=chi2,k=5)
```

```python
#train the model
trf5 = DecisionTreeClassifier()
```

## create pipeline

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```
